```python
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.cm import get_cmap
import numpy as np
from scipy.special import lambertw
import csv
import os
import re
from os import listdir
from os.path import isfile, join

def clean_raw_data(iv_folder, iv_file, vmin, vmax):
    
    """Cleans a txt file with voltage and current data."""
    
    data = np.loadtxt(iv_folder + "/" + iv_file, usecols=(0, 1))
    
    # Find the indices of entries where voltage is 0
    zero_voltage_indices = np.where(data[:, 0] == 0)[0]

    # If there are duplicates at voltage 0, remove one of each pair
    if len(zero_voltage_indices) > 1:
        # Remove the first duplicate (assuming they are consecutive)
        data = np.delete(data, zero_voltage_indices[0], axis=0)

    # Sort the data from reverse to forward bias (sort by voltage in ascending order)
    truncated_data = data[(data[:, 0] >= vmin) & (data[:, 0] <= vmax)]
    sorted_data_reverse_to_forward = truncated_data[truncated_data[:, 0].argsort()]
    
    # Separate the re-sorted data into voltage and current arrays
    voltage = sorted_data_reverse_to_forward[:, 0]
    current = sorted_data_reverse_to_forward[:, 1]
    
    return voltage, current
    
def load_result_file(results_folder, results_file):
    """Load a txt file with results from several optimization runs"""
    df = pd.read_csv(results_folder + "/" + results_file, delim_whitespace=True, header=0, dtype=float)
    df.set_index('Run', inplace=True)
    df.index = df.index.astype(int)
    return df   

def calculate_mape(actual, forecast):    
    n = len(actual)
    mape = 100*sum(abs((a - f) / a) for a, f in zip(actual, forecast))/n
    return mape

def lambertw_newton(z, tol=1e-10, max_iter=100):
    w = np.log(z + 1)  # Initial guess
    for i in range(max_iter):
        ew = np.exp(w)
        w_next = w - (w * ew - z) / (ew + w * ew)
        if np.all(np.abs(w_next - w) < tol):
            return w_next
        w = w_next
    return w  # Return the last approximation if convergence criteria are not met

def parse_h2_flow_ppm(filename):
    # Use regex to find the H2 flow rate directly
    try:
        h2_flow_match = re.search(r'H2_(\d+)sccm', filename)
        if h2_flow_match:
            forming_gas_flow = float(h2_flow_match.group(1))
            # Calculate the actual hydrogen content in ppm
            h2_concentration_ppm = (2 / 100) * forming_gas_flow * 1e6 / 100  # 2% of forming gas, normalized to 100 sccm total flow
            return h2_concentration_ppm
        else:
            return "unknown"
    except ValueError as e:
        print(f"Error parsing H2 flow from {filename}: {e}")
        return "unknown"

def lighten_color(color, amount=0.5):
    """ Lightens the given color by multiplying (1-luminosity) by the given amount.
        Input can be matplotlib color string, hex string, or RGB tuple.

        Examples:
        - lighten_color('g', 0.3)
        - lighten_color('#F034A3', 0.6)
        - lighten_color((0.3, 0.55, 0.1), 0.5)
    """
    import matplotlib.colors as mc
    import colorsys
    try:
        c = mc.cnames[color]
    except:
        c = color
    c = colorsys.rgb_to_hls(*mc.to_rgb(c))
    return colorsys.hls_to_rgb(c[0], 1 - amount * (1 - c[1]), c[2])

def results(temperature_celsius, iv_folder, iv_files, results_folder, results_files, plot=False, save=False):
    area = 3.14 * (0.1 / 2) ** 2
    charge = 1.60e-19
    boltz = 1.38e-23
    vt = boltz * (temperature_celsius + 273.15) / charge
    
    # Get maximum H2 concentration for color scaling
    h2_ppm_values = [parse_h2_flow_ppm(iv_file) for iv_file in iv_files]
    max_h2_ppm = max(h2_ppm_values)
    cmap = get_cmap('coolwarm')  # Color map from cold (blue) to hot (red)
    
    if plot:
        plt.figure(figsize=(8, 6))

    for i, (iv_file, results_file) in enumerate(zip(iv_files, results_files)):
        voltage, current = clean_raw_data(iv_folder, iv_file, vmin, vmax)
        df = load_result_file(results_folder, results_file)
        best_result = df.loc[df['Objective'].idxmin()]
        objective, i0, n, Rs, Rsh, offset = best_result
        indv = [i0, n, Rs, Rsh, offset]
        print(f"Parameters from {results_file}: {best_result}")

        h2_ppm = h2_ppm_values[i]
        label = "H\u2082: trace" if h2_ppm == 0 else f"H\u2082: {int(h2_ppm)} ppm"

        #color = cmap(h2_ppm / max_h2_ppm)  # Normalize H2 concentration for color map
        color='firebrick'
        lighter_color = lighten_color(color, amount=0.5)

        adjusted_voltage = voltage + indv[4]
        term1 = 1.0 + indv[2] / indv[3]
        term2 = (indv[0] * indv[2]) / (indv[1] * vt * term1)
        term3 = (indv[2] * indv[0] + adjusted_voltage) / (indv[1] * vt * term1)
        sim = (indv[1] * vt / indv[2]) * lambertw_newton(term2 * np.exp(term3)) - (indv[0] - adjusted_voltage / indv[3]) / term1
        mape = calculate_mape(sim, current)
        print(f"MAPE for {iv_file}: {mape:.2f} %")

        if plot:
            plt.plot(voltage, np.abs(current / area), 'o', color=lighter_color, label=f"{label}")
            plt.plot(voltage, np.abs(sim / area), '-', color=color, label=f"Fit")

    if plot:
        plt.yscale("log")
        plt.xlabel('Voltage (V)')
        plt.ylabel('Current density ($A\,cm^{-2}$)')
        plt.legend(loc='upper left', frameon=False)
        if save:
            plt.savefig('./Figures/IV_fit.png', format='png', dpi=600, bbox_inches='tight')
        plt.show()
```

```python
# %matplotlib qt 
plt.rcParams['font.size'] = 16
pd.set_option('display.float_format', lambda x: '%.3e' % x)
temperature_celsius = 500
vmin, vmax = -10, 5

iv_folder = "./D2_Gas/500C"
results_folder = iv_folder + "/Results"


iv_files = ['iv_curve_500.0_N2_100sccm_H2_0sccm_2024-03-29_13-06-11.txt']
            
results_files = ['Results_iv_curve_500.0_N2_100sccm_H2_0sccm_2024-03-29_13-06-11.txt']


results(temperature_celsius, iv_folder, iv_files, results_folder, results_files, plot=True, save=False)
```

```python
import os
import re
import pandas as pd
import matplotlib.pyplot as plt

# Function to load data from files
def load_results(folder_path):
    # Define regex patterns for extracting H2 flow and timestamp
    h2_flow_pattern = r"H2_(\d+)sccm"
    timestamp_pattern = r"(\d{4}-\d{2}-\d{2}_\d{2}-\d{2}-\d{2})"
    
    # List to hold each best row DataFrame
    best_rows = []
    
    # Iterate over each file in the specified folder
    for filename in os.listdir(folder_path):
        if filename.endswith(".txt"):
            # Extract H2 flow and timestamp from the file name
            h2_flow = int(re.search(h2_flow_pattern, filename).group(1))
            timestamp = re.search(timestamp_pattern, filename).group(1)
            
            # Load the file contents into a pandas DataFrame
            file_path = os.path.join(folder_path, filename)
            df = pd.read_csv(file_path, sep='\s+')
            
            # Find the row with the lowest objective
            best_row = df.loc[[df['Objective'].idxmin()]]
            
            # Add the extracted H2 flow and timestamp as new columns
            best_row['H2_Flow (sccm)'] = h2_flow
            best_row['Timestamp'] = pd.to_datetime(timestamp, format='%Y-%m-%d_%H-%M-%S')
            
            # Append the best row DataFrame to the list
            best_rows.append(best_row)
    
    # Concatenate all best row DataFrames from the list into one DataFrame
    best_results_df = pd.concat(best_rows, ignore_index=True)
    
    return best_results_df

# Load the data
folder_path = "./D2_Gas/500C/Results"
gas_df = load_results(folder_path)
#gas_df
```

```python
# Function to plot parameter vs. time elapsed for specific H2 flows
plt.rcParams['font.size'] = 18
def plot_vs_time_elapsed_for_flow(ax, df, parameter, flow_value, plot_color):
    area = 3.14 * (0.1 / 2) ** 2  # cm^2
    # Convert flow value to ppm
    #flow_value_ppm = 1e6 * 0.02 * flow_value / 100
    flow_value_ppm = flow_value
    
    # Filter the DataFrame for the specified H2 flow
    filtered_df = df[df['H2_Flow (sccm)'] == flow_value].copy()
    
    # Calculate the time elapsed from the first data point
    start_time = filtered_df['Timestamp'].iloc[0]
    filtered_df['Time Elapsed'] = (filtered_df['Timestamp'] - start_time).dt.total_seconds() / 60  # in minutes
           
    # Plotting
    if parameter == "I0":
        marker = 'o' if flow_value == 0 else ('s' if flow_value == 2 else '^')
        ax.plot(
            filtered_df['Time Elapsed'],
            1000 * filtered_df[parameter] / area,
            marker=marker,
            markersize=10,  # Increase marker size
            markerfacecolor=plot_color,
            markeredgecolor='black',  # Add black border
            linestyle='None',  # No lines connecting markers
            label=f'{int(flow_value_ppm)} % H/N' if flow_value > 0 else 'Trace H/N'
        )
        
     
    elif parameter == "Rs" or parameter == "Rsh":
        marker = 'o' if flow_value == 0 else ('s' if flow_value == 2 else '^')
        ax.plot(
            filtered_df['Time Elapsed'],
            filtered_df[parameter]*area,
            marker=marker,
            markersize=10,  # Increase marker size
            markerfacecolor=plot_color,
            markeredgecolor='black',  # Add black border
            linestyle='None',  # No lines connecting markers
            label=f'{int(flow_value_ppm)} % H/N' if flow_value > 0 else 'Trace H/N'
        )
        
    elif parameter == "n":
        marker = 'o' if flow_value == 0 else ('s' if flow_value == 2 else '^')
        ax.plot(
            filtered_df['Time Elapsed'],
            filtered_df[parameter],
            marker=marker,
            markersize=10,  # Increase marker size
            markerfacecolor=plot_color,
            markeredgecolor='black',  # Add black border
            linestyle='None',  # No lines connecting markers
            label=f'{int(flow_value_ppm)} % H/N' if flow_value > 0 else 'Trace H/N'
        )             
    
# Create a 2x2 subplot grid
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot for J0
plot_vs_time_elapsed_for_flow(axs[0, 0], gas_df, "I0", 0, 'royalblue')
plot_vs_time_elapsed_for_flow(axs[0, 0], gas_df, "I0", 2, 'lightblue')
plot_vs_time_elapsed_for_flow(axs[0, 0], gas_df, "I0", 5, 'tomato')
axs[0, 0].set_ylabel('J (mA cm)')
axs[0, 0].set_ylim(0, 1.2)  # Limit x-axis to 0-20 minutes
axs[0, 0].legend(loc='upper left', frameon=False, fontsize='16',title_fontsize='16')  # Only J0 with legend, no frame

# Plot for n
plot_vs_time_elapsed_for_flow(axs[0, 1], gas_df, "n", 0, 'royalblue')
plot_vs_time_elapsed_for_flow(axs[0, 1], gas_df, "n", 2, 'lightblue')
plot_vs_time_elapsed_for_flow(axs[0, 1], gas_df, "n", 5, 'tomato')
axs[0, 1].set_ylabel('Ideality factor')
axs[0, 1].set_ylim(1.1, 1.6)  

# Plot for Rs
plot_vs_time_elapsed_for_flow(axs[1, 0], gas_df, "Rs", 0, 'royalblue')
plot_vs_time_elapsed_for_flow(axs[1, 0], gas_df, "Rs", 2, 'lightblue')
plot_vs_time_elapsed_for_flow(axs[1, 0], gas_df, "Rs", 5, 'tomato')
axs[1, 0].set_ylabel('R ( cm)')
axs[1, 0].set_ylim(0.1, 0.4) 

# Plot for Rsh
plot_vs_time_elapsed_for_flow(axs[1, 1], gas_df, "Rsh", 0, 'royalblue')
plot_vs_time_elapsed_for_flow(axs[1, 1], gas_df, "Rsh", 2, 'lightblue')
plot_vs_time_elapsed_for_flow(axs[1, 1], gas_df, "Rsh", 5, 'tomato')
axs[1, 1].set_ylabel('R ( cm)')
axs[1, 1].set_ylim(380, 400) 

# Finalizing the plot
for ax_row in axs:
    for ax in ax_row:
        ax.set_xlabel('Time Elapsed (minutes)')
        ax.grid(False)  # Remove gridlines

plt.tight_layout()  # Adjust subplot layout to prevent overlapping
plt.show()  # Display the plot
```
